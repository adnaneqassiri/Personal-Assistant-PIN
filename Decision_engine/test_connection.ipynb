```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, from_json, struct, coalesce
from pyspark.sql.types import *

# -------------------------------
# 1. Spark Session
# -------------------------------
spark = SparkSession.builder \
    .appName("App") \
    .master("spark://localhost:7077") \
    .config("spark.driver.host", "172.21.0.1") \
    .config("spark.driver.bindAddress", "0.0.0.0") \
    .config("spark.driver.port", "41251") \
    .config("spark.blockManager.port", "41252") \
    .config("spark.executor.memory", "1g") \
    .config("spark.executor.cores", "1") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")
```

```python
print(spark.sparkContext.master)
print(spark.sparkContext._jsc.sc().getExecutorMemoryStatus().keySet().toString())

spark.range(10).show()
```
